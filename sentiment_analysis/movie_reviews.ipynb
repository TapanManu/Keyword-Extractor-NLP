```python
from nltk.tokenize import sent_tokenize
text="Hello all! I am  really satisfied with his performance.His attitude was really good and awesome."
tokenized_text=sent_tokenize(text)
tokenized_text
```

```python
from nltk.tokenize import word_tokenize
tokenized_word=word_tokenize(text)
tokenized_word
```

```python
from nltk.probability import FreqDist
fdist=FreqDist(tokenized_word)
fdist
```

```python
fdist.most_common(4)
```

```python
import matplotlib.pyplot as plt
fdist.plot(30,cumulative=False)
plt.show()
```

```python
from nltk.corpus import stopwords
stop_words=set(stopwords.words("english"))
stop_words
```

```python
filtered_sent=[w for w in tokenized_word if w not in stop_words]
filtered_sent
```

```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()
stemmed_words=[ps.stem(w) for w in filtered_sent]
stemmed_words
```

```python
from nltk.stem.wordnet import WordNetLemmatizer
lem=WordNetLemmatizer()
word='are'
print('lemmatized word:',lem.lemmatize(word,'v'))
print('stemmed word:',ps.stem(word))
```

```python
import pandas as pd
movie_review = pd.read_csv("../input/movie-review/movie_review.csv")
movie_review.head()
```

```python
movie_review['tag']=movie_review['tag'].apply(lambda x:1 if x=='pos' else 0)
movie_review.tail()
```

```python
movie_review.tag.value_counts()
```

```python
import seaborn as sns
Sent_count=movie_review.groupby('tag').count()
Sent_count
```

```python
sns.barplot(x=Sent_count.index.values,y=Sent_count['text'])
plt.xlabel('Review of statements')
plt.ylabel('Number of review')
plt.show()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from nltk.tokenize import RegexpTokenizer
#tokenizer to remove unwanted elements from out data like symbols and numbers
token = RegexpTokenizer(r'[a-zA-Z0-9]+')
cv = CountVectorizer(lowercase=True,stop_words='english',ngram_range = (1,1),tokenizer = token.tokenize)
text_counts= cv.fit_transform(movie_review['text'])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    text_counts, movie_review['tag'], test_size=0.3, random_state=1)
```

```python
from sklearn.naive_bayes import MultinomialNB
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
# Model Generation Using Multinomial Naive Bayes
clf = MultinomialNB().fit(X_train, y_train)
predicted= clf.predict(X_test)
print("MultinomialNB Accuracy:",metrics.accuracy_score(y_test, predicted))
```
